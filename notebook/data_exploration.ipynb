# Data Exploration — Sweetviz Report\n\nAnálise exploratória do dataset `kc_house_merged.csv` usando Sweetviz.

```python
import pandas as pd
import sweetviz as sv

# Carregar dados
df = pd.read_csv("../data/kc_house_merged.csv")
print(f"Shape: {df.shape}")
df.head()
```

```python
# Gerar relatório Sweetviz com price como target
report = sv.analyze(df, target_feat="price")

# Salvar como HTML e exibir no notebook
report.show_html("sweetviz_report.html", open_browser=False)
report.show_notebook()
```
